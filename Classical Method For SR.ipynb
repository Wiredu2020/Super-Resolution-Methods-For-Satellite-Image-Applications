```python
import cv2
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
def RGB_conv(image, kernel):
    result = np.empty_like(image)
    for dim in range(result.shape[-1]):
        result[:,:,dim] = cv2.filter2D(image[:,:,dim],-1,kernel)
        #result[:,:,dim] = signal.convolve2d(image[:,:,dim],kernel,mode="same",boundary="fill")
        result[:,:,dim] = np.clip(result[:,:,dim], 0, 255)
    return result
```

```python
img = cv2.imread('moon.png', 0)
img2 = cv2.imread('sat1.jpg')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
#Laplacian
kernel1 = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]], dtype=np.float32)

#Variant Laplacian
kernel2 = np.array([[-1, -1, -1], 
                    [-1, 9, -1], 
                    [-1, -1, -1]]) 
result1 = cv2.filter2D(img,-1,kernel1)

result2 = cv2.filter2D(img,-1,kernel2)
result3 = RGB_conv(img2,kernel1)

result4 = RGB_conv(img2,kernel2)

fig, axs = plt.subplots(1, 3, figsize=(10, 5))
axs[0].imshow(img, cmap='gray')
axs[0].set_title('Original')
axs[0].axis('off')

axs[1].imshow(result1, cmap='gray')
axs[1].set_title('Laplacian Filter')
axs[1].axis('off')

axs[2].imshow(result2, cmap='gray')
axs[2].set_title('Varaint Laplacian Filter')
axs[2].axis('off')


plt.subplots_adjust(wspace=0.01)
plt.subplots_adjust(hspace=0)

plt.show()
```
