```python
from PIL import Image
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
image = Image.open('Data/Test/HR/baboon.bmp')
```

```python
img = cv2.imread('Data/Test/HR/baboon.bmp')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
img
```

```python
plt.imshow(img)
plt.show()
```

```python
#Taking part of image
img_part = img[10:100,10:100,:]
plt.imshow(img_part)
plt.show()
```

```python
start_point = (10, 10) 
  
# End coordinate, here (250, 250) 
# represents the bottom right corner of image 
end_point = (100, 100) 
  
# Green color in BGR 
color = ( 255,0, 0) 
  
# Line thickness of 9 px 
thickness = 2
  
# Using cv2.line() method 
# Draw a diagonal green line with thickness of 9 px 
im_rec = cv2.rectangle(img, start_point, end_point, color, thickness) 
plt.imshow(im_rec)
plt.show() 
```
